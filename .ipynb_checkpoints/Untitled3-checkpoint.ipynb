```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/train.csv')
```

```python
#return all numeric columns 
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numeric = df.select_dtypes(include=numerics)
numeric.columns
```

```python
categorical = df.select_dtypes(exclude=numerics)
```

```python
def group_low_frequency_categories(x):
    cols = x.columns.values
    for col in cols:
        counts= x[col].value_counts()
        length = x[col].count()
        threshold = length * 0.05
        values = x[col].unique()
        for value in values:
            if counts[value] < threshold:
                print(value)
                x[col].replace(value,'Other',inplace=True)
```

```python
categorical = categorical.replace(np.nan, 'null_value', regex=True)
```

```python
group_low_frequency_categories(categorical)
```

```python
cols = categorical.columns.values
categorical['income_level'] =df['income_level']
for col in cols:
    plt.figure(col)
    sns.countplot(y=col, hue="income_level", data=categorical)
```

```python
train['income_level'] = np.where(train['income_level'] == -50000,0,1)
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for col in train.columns.values:
    # Encoding only categorical variables
    if train[col].dtypes=='object':
        # Using whole data to form an exhaustive list of levels
        print(col)
        data=train[col]
        le.fit(data.values.astype(str))
        train[col]=le.transform(data.astype(str))
```

```python
train.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale

pca = PCA(n_components=37)
pca.fit(train)
```

```python
#The amount of variance that each PC explains
var= pca.explained_variance_ratio_

#Cumulative Variance explains
var1=np.cumsum(np.round(pca.explained_variance_ratio_, decimals=4)*100)
```

```python
plt.plot(var1)
```

```python
pca = PCA(n_components=25)
pca.fit_transform(train)
```

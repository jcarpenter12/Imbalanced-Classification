```python
def group_low_frequency_categories(x):
    cols = x.columns.values
    for col in cols:
        print("column name %s" % col)
        counts= x[col].value_counts()
        length = x[col].count()
        threshold = length * 0.05
        print("length: %s" % length)
        print("threshold: %s" % threshold)
        values = categorical[col].unique()
        for value in values:
            if counts[value] < threshold:
                print(value)
                x[col].replace(value,'Other',inplace=True)

major_industry = pd.DataFrame(categorical['major_industry_code'])
print(major_industry['major_industry_code'].value_counts())
group_low_frequency_categories(major_industry)


```

```python
new = categorical.drop('income_level',1)
group_low_frequency_categories(categorical)
```

```python
#Define a generic function using Pandas replace function
def coding(col, codeDict):
    colCoded = pd.Series(col, copy=True)
    for key, value in codeDict.items():
        colCoded.replace(key, value, inplace=True)
    return colCoded
 
```
